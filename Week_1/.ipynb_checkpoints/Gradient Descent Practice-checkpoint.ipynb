Gradient Descent Normal

```python
def f_x(x):
    ret=x**4-3*x**3+2
    return ret
```

```python
def f_prime_x(x):
    ret=4*x**3-9*x**2
    return ret
```

```python
def gradient_descent():
    x_old=0.0
    x_new=6.0
    precission=0.00001
    max_iter=1000
#     learning_rate = 0.01
    learning_rate=0.0001
#     learning_rate = 0.1
    iter=0
    
    while abs(x_new-x_old)>precission and iter < max_iter:
        x_old=x_new
        x_new=x_old-learning_rate*f_prime_x(x_old)
        iter+=1
        print('iter',iter,'new_x',x_new)
    
    print('Local Minimum at', x_new)
    print('Function Value', f_x(x_new))
```

```python
gradient_descent()
```

Gradient Descent Brute Force

```python
def bruto_force():
    x_start=-10.0
    x_end=10.0
    max_iter=1000
#     learning_rate=0.01
#     learning_rate=0.0001
    learning_rate=0.1
    iter=0
    min_x=x_start
    min_f_x=f_x(min_x)
    x=x_start
    
    while x<x_end and iter<max_iter:
        new_f_x=f_x(x)
        if new_f_x<min_f_x:
            min_x=x
            min_f_x=new_f_x
        x+=learning_rate
        iter+=1
        print('iter',iter,'new_x',min_x)
        
    print('Local Minimum at', min_x)
    print('Function Value', min_f_x)
```

```python
bruto_force()
```
